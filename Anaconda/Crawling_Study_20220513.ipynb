```python
#네이버 실시간 정치뉴스 크롤링
import urllib.request
import bs4
import csv

list=[]
idx=1
url="https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=100"
html=urllib.request.urlopen(url)

soup=bs4.BeautifulSoup(html,"html")
hdline=soup.find("ul",{"class":"section_list_ranking_press _rankingList"})
lis=hdline.findAll("li")
for li in lis:
    temp=[]
    a=li.find("a",{"class":"list_tit nclicks('rig.renws2')"})
    title=idx,a.text.strip()
    urla=a.get('href')
    print(title)
    print(urla)
    print()
    temp.append(idx)
    temp.append(title)
    temp.append(urla)
    list.append(temp)
    idx+=1
f=open("Hdline.csv","w",encoding="cp949",newline="")#엔터 방지용 newline
header=["NO","Title","Url"]
csvwriter=csv.writer(f)
csvwriter.writerow(header)
for i in list:
    csvwriter.writerow(i)
f.close()
```

```python
#가우스전가(3~4) 웹툰정보 크롤링 예제
import requests
from bs4 import BeautifulSoup

url="https://comic.naver.com/webtoon/list?titleId=675554"
html=requests.get(url)
soup=BeautifulSoup(html.text,"html")

#만화제목과 링크 가져오기
idx=1
total_rate=0
cartoons=soup.findAll("td",{"class":"title"})
for cartoon in cartoons:
    title=cartoon.a.get_text()
    link="https://comic.naver.com"+cartoon.a.get('href')
    print(idx,title,link)
    idx+=1
    
#평점구하기
cartoons=soup.findAll("div",{"class":"rating_type"})
for cartoon in cartoons:
    rate=cartoon.find("strong").get_text()
   # print(rate)
    total_rate+=float(rate)
print("총점수 :",total_rate)
print("평균 :",total_rate/len(cartoons))
```

```python
#정규식 : 주민번호,차량번호,이메일 주소
import re
p=re.compile("ca.e")
# .:하나의 문자를 으미
# ^: 문자열의 시작
# $문자열의 끝

m=p.match("cafe")
def print_match(m):
    if m:
        print("m.group()",m.group()) #일치하는 문자열을 반환
        print("m.string",m.string)#압력받은 문자열
        print("m.start()",m.start())#일치하는 문자열의 시작
        print("m.end()",m.end()) #일치하는 문자열의 끝 인덱스
        print("m.span()",m.span())#일치하는 문자열의 끝과 시작의 인덱스
    else:
        print("매치되지 않았습니다")
m=p.match("careless")
lst=p.findall("good care cafe case")
print_match(m)
print(lst)
```

```python
#이미지 파일을 크롤링하여 저장하는 내용임
#다음 영화를 검색하고 연순위별 5개 이미지를 다운로드
#최근 5년 동안의 인기있는 영화를 골라 저장한다.
import requests #from urllib, request import urlopen
from bs4 import BeautifulSoup
url="https://search.daum.net/search?w=tot&q=2021%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"
res=requests.get(url)
res.raise_for_status()#정상연결 유무를 검증, 연결시 통과
soup=BeautifulSoup(res.text,"html")

images=soup.findAll("img",{"class":"thumb_img"})
for idx,image in enumerate(images): #enumerate() 번호를 할당하여 구성을 시켜주는 함수
    image_url=image["src"]
#     print(image_url)
    if image_url.startswith("//"):
        image_url="https:"+image_url
        
    print(image_url)    
#접속해서 파일을 저장
    image_res=requests.get(image_url) #이미지의 url을 넘겨주어 접속
    image_res.raise_for_status()
    with open("movie{}.jpg".format(idx+1),"wb")as f: #여기에 movie1,2,3...
        f.write(image_res.content) #image가 곧 content
    
    #상위 5개 이미지만
    if idx>=4:
        break
```

```python
#이미지 파일을 최근 5년동안 연도별로 저장하는 내용임

import requests #from urllib, request import urlopen
from bs4 import BeautifulSoup

for year in range(2017,2022):
    url="https://search.daum.net/search?w=tot&q={}%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR".format(year)
    res=requests.get(url)
    res.raise_for_status()#정상연결 유무를 검증, 연결시 통과
    soup=BeautifulSoup(res.text,"html")

    images=soup.findAll("img",{"class":"thumb_img"})

    for idx,image in enumerate(images): #enumerate() 번호를 할당하여 구성을 시켜주는 함수
        image_url=image["src"]
    #     print(image_url)
        if image_url.startswith("//"):
            image_url="https:"+image_url

        print(image_url)    
    #접속해서 파일을 저장
        image_res=requests.get(image_url) #이미지의 url을 넘겨주어 접속
        image_res.raise_for_status()
        with open("movie{}_{}.jpg".format(year,idx+1),"wb")as f: #여기에 movie1,2,3...
            f.write(image_res.content) #image가 곧 content

        #상위 5개 이미지만
        if idx>=4:
            break
```

```python
#코스피시가총액순위를 네이버에서 검색해서 가져와보자
import csv
import re
import requests
from bs4 import BeautifulSoup

url="https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page="
for page in range(1,2): #1페이지만 선택한경우 (50위까지의 데이터)
    res=requests.get(url+str(page)) #str 감싸줌
    res.raise_for_status()
    soup=BeautifulSoup(res.text,"html")
    
    #파싱을 하면 테그가 table과 클래스 type_2.. 그 밑의 today 그밑의 tr그밑의 td 구성
    data_rows=soup.find("table",{"type_2"}).find("tbody").findAll("tr")
    for row in data_rows:
        columns=row.findAll("td")
        if len(columns)<=1: #의미없는 데이터를 고르는 내용
            continue
        data=[column.get_text().strip() for column in columns] #1줄짜리 for문으로 컬럼스 안에 컬럼에 대한 컬럼의 내용들을
        #추출하는 내용
        print(data)
```
